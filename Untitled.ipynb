```python
class Animal(object):
    def __init__(self, speed, is_mammal):

        self.speed = speed
        self.is_mammal = is_mammal

class Cat(Animal):
    def __init__(self, is_hungry):
        super(Cat,self).__init__(10, True)
        self.is_hungry = is_hungry

barry = Cat(False)
print(f"speed: {barry.speed}")
print(f"is a mammal: {barry.is_mammal}")
print(f"feed the cat?: {barry.is_hungry}")
```

```python
class Person( object ):     
  
    # __init__ is known as the constructor          
    def __init__(self, name, idnumber):    
        self.name = name 
        self.idnumber = idnumber 
    def display(self): 
        print(self.name) 
        print(self.idnumber) 
  
 
class Employee( Person ):            
    def __init__(self, name, idnumber, salary, post): 
        self.salary = salary 
        self.post = post 

        # invoking the __init__ of the parent class
        # Person.__init__(self, name, idnumber)
        super(Employee,self).__init__(name,idnumber)
        super().__init__(name,idnumber)
    
    def display_child(self):
        super().display()
    
    # overwrite parent function
    def display(self):
        print(self.name) 
  
                  

a = Employee('Rahul', 886012, 200000, "Intern")     
  
# calling a function of the class Person using its instance 
a.display() 
a.display_child()
```
